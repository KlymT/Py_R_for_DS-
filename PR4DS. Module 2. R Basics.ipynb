## Loop

### For

```
for (variable in sequence){
   Do something
}
while (condition){
   Do something
}
```

````
for (i in 1:4){ 
  j <- i + 10
  if(j == 11) {
    next
  }
  if(j == 13) {
    break
  }
  print(j) 
}
````

### while

```
while (condition){
   Do something
}
```

````
i = 0
while (i < 5){
   print(i)
   i <- i + 1 
  }
````

### if else

```
if (condition){
   Do something
} else {
   Do something different
}
```

````
i <- 7

if (i > 3){ 
  print("Yes")
} else { 
  print("No")
}
````

````
ifelse(df[['votes']] > 225, "Yes", "No")
````
