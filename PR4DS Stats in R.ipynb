```
df = readr::read_csv("https://raw.githubusercontent.com/alexlorenzo/Pet.my/master/datasets/train.csv")

df
```

## Тести на перевірку нормальності розподілу

```
shapiro.test(df$Age[1:500])
```

```
install.packages("moments")
```

```
moments::agostino.test(df$Age)
```

```
install.packages("nortest")
```

```
nortest::ad.test(df$Age)
```

## Тести на кореляцію

```
cor(dplyr::select_if(df, is.numeric), method = "kendall")
```

```
cor.test(df$Age, df$Health)
```

```
cor.test(df$Age, df$Health, method = "kendall")
```

```
cor.test(df$Age, df$Health, method = "spearman")
```
